## Q1: Are more unique models using alternative sources of fuel? By how much?

### 1) models using alternative sources of fuel in 2008

```python
df_08.fuel.value_counts()
```

```python
model_alternate_fuels_08 = df_08.query("fuel == 'CNG' or fuel == 'ethanol'")
```

```python
print(model_alternate_fuels_08.fuel.value_counts())
alt_08 = model_alternate_fuels_08.model.nunique()
print(alt_08)
```

### 2) models using alternative sources of fuel in 2018

```python
df_18.fuel.value_counts()
```

```python
model_alternate_fuels_18 = df_18.query('fuel in ["Ethanol", "Electricity"]')
```

```python
print(model_alternate_fuels_18.fuel.value_counts())
alt_18 = model_alternate_fuels_18.model.nunique()
print(alt_18)
```

```python
plt.bar(["2008", "2018"], [alt_08, alt_18])
plt.title("Number of Unique Models Using Alternative Fuels")
plt.xlabel("Year")
plt.ylabel("Number of Unique Models");
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokeman = pd.read_csv('pokemon.csv')
```

```python
print(pokeman.shape)
pokeman.head()
```

```python
pkmn_stats = ['hp', 'attack', 'defense', 'speed', 'special-attack', 'special-defense']
g = sb.PairGrid(data=pokeman, vars = pkmn_stats);
g = g.map_offdiag(plt.scatter);
g.map_diag(plt.hist);
```

```python
pokeman[pkmn_stats].corr()
```

```python
sb.heatmap(pokeman[pkmn_stats].corr(), cmap='rocket_r', annot=True, fmt='.2f');
```

```python
sb.heatmap(pokeman[pkmn_stats].corr(), cmap='rocket_r', annot=True, fmt='.2f', vmin=0, center=0);
```
